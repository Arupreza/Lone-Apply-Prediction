```python
#Input After May moth data for predicting result
loan_result = pd.read_csv('df_2022_06.csv')
loan_result
```

```python
#Input result data
result = pd.read_csv('데이터분석분야_퓨처스부문_평가데이터.csv')
result = result[['application_id', 'product_id']]
result
```

```python
#Marge features with result data
result_final_submit = pd.merge(result, loan_result, on=["application_id", "product_id"])
result_final_submit
```

```python
#result_final_submit.to_csv('result_final_submit.csv', index=False)
```

```python
#Feature selection
col = ['bank_id', 'product_id', 'loan_limit', 'loan_rate', 'birth_year',
       'gender', 'credit_score', 'yearly_income', 'income_type_labeled',
       'employment_type_labeled', 'houseown_type_labeled', 'purpose_labeled',
       'desired_amount', 'personal_rehabilitation_yn',
       'personal_rehabilitation_complete_yn', 'existing_loan_cnt',
       'existing_loan_amt', 'GetCreditInfo', 'UseLoanManage', 'Login',
       'EndLoanApply', 'OpenApp', 'ViewLoanApplyIntro', 'StartLoanApply',
       'CompleteIDCertification', 'UsePrepayCalc', 'SignUp', 'UseDSRCalc']
```

```python
result_final_submit_scale = result_final_submit[col]
```

```python
#Data Scaling
scaler = MinMaxScaler(feature_range=(0, 1))
result_final_submit_scale = scaler.fit_transform(result_final_submit_scale)
```

```python
#Input selection
result_final_submit_scale = result_final_submit_scale[:, :28]
```

```python
#Resize data
result_final_submit_scale_input = np.reshape(result_final_submit_scale, (result_final_submit_scale.shape[0], 1, 28, 1))
```

```python
#Input trained model
from keras.models import load_model

# returns a compiled model
# identical to the previous one
model_1 = load_model('Final_Project_CNN_1.h5')
model_2 = load_model('Final_Project_CNN_2.h5')
model_3 = load_model('Final_Project_CNN_3.h5')
```

```python
#Join all model
from keras.models import Sequential
import tensorflow as tf
from tensorflow.keras.models import Model
m1 = load_model('Final_Project_CNN_1.h5')
m1._name = 'test1'
m2 = load_model('Final_Project_CNN_2.h5')
m2._name = 'test2'
m3 = load_model('Final_Project_CNN_3.h5')
m3._name = 'test3'
commonInput = tf.keras.Input(shape=(1, 28, 1))
out1 = m1(commonInput)
out2 = m2(commonInput)
out3 = m2(commonInput)
merged = tf.keras.layers.Add()([out1,out2,out3])
newModel = Model(commonInput, merged)
```

```python
#Prediction
pro_result = newModel.predict(result_final_submit_scale_input)
```

```python
#Convert categorical prediction to result format
pro_result = pd.DataFrame(np.argmax(pro_result, axis=1).astype(int), columns=['is_applied'])
```

```python
#Joind the result with feature
Final_output = pd.concat([result_final_submit, pro_result], axis = 1)
Final_output
```
